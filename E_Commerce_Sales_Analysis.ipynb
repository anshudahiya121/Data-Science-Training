```python
import pandas as pd
df=pd.read_csv('data.csv',encoding='ISO-8859-1')
df.head()
```

```python
df.dropna(subset=["CustomerID"], inplace=True)
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df=df[df['Quantity']>0]
df['TotalRvenue']=df['Quantity']*df['UnitPrice']
print("Data is Cleaned")
```

```python
best_seller=df.groupby('Description')['Quantity'].sum().reset_index()
best_seller=best_seller.sort_values('Quantity',ascending=False)
print(best_seller.head(10))
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_revenue = df.groupby('Month')['TotalRvenue'].sum().reset_index()
print(monthly_revenue)
```

```python
customer_habits=df.groupby('CustomerID').agg({'InvoiceNo':'nunique','TotalRvenue':'sum'}).reset_index()
customer_habits.columns=['CustomerID','TotalOrders','TotalSpent']
print(customer_habits.sort_values('TotalSpent',ascending=False).head(10))
```

```python
all_money=df['TotalRvenue'].sum()
all_orders = df['InvoiceNo'].nunique()
avg_order_value = all_money / all_orders
print("Total Money Made:", all_money)
print("Total Orders Placed:", all_orders)
print("Average Money Spent Per Order:", avg_order_value)

```
